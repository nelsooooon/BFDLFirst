# Import Library

```python
from google_play_scraper import Sort, reviews

import pandas as pd
```

# Scraping Dataset

```python
# Mengambil semua ulasan dari aplikasi dengan ID 'com.google.android.apps.classroom' di Google Play Store.
scrapreview = reviews(
    'com.google.android.apps.classroom',
    lang='en',
    country='us',
    sort=Sort.MOST_RELEVANT,
    count=50000 # 50 ribu sampel data
)
```

# Saving Dataset


```python
# Membuat DataFrame dari hasil scrapreview
app_reviews_df = pd.DataFrame(scrapreview[0])

# Menyimpan ulasan dalam file CSV
app_reviews_df.to_csv('ulasan_classroom.csv', index=False)
```
